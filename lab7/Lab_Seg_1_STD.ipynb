```python
req_2=io.imread("imgs/exp2/cufe.png").astype(float)
io.imshow(req_2/255)
```

```python
G=req_2[:,:,1]
io.imshow(G,cmap=mpl.colormaps["Greens"])
```

```python
B=req_2[:,:,2]
io.imshow(B,cmap=mpl.colormaps["Blues"])
```

```python
R_=R-72
G_=G-95
B_=B-35
```

```python
img_f=np.dstack((R_, G_, B_))
```

```python
ff=np.linalg.norm(img_f, axis=2)
```

```python
mask=ff<34
io.imshow(mask)
```
