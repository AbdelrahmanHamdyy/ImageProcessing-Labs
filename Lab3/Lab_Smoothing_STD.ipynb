```python
from commonfunctions import *
import numpy as np
from skimage.filters import median
from skimage.filters import gaussian
import skimage.io as io
import math

%matplotlib inline
%load_ext autoreload
%autoreload 2
```

```python
img = io.imread('./pout.tif', as_gray=True)
noisy_img = random_noise(img, mode="s&p", amount = 0.05)
```

```python
io.imshow(img)
```

```python
io.imshow(noisy_img)
```

```python
'''
(1) Your own median filter algorithm
'''
filter_size = 3

width = img.shape[0]
height = img.shape[1]

d = filter_size//2

output_img = np.zeros_like(noisy_img)

for i in range(d, width - d):
        for j in range(d, height - d):
            median_filter = np.copy(noisy_img[i: i + filter_size, j: j + filter_size])
            output_img[i, j] = np.median(median_filter)

io.imshow(output_img)
```

```python
'''
(2) skimage median filter
'''
med_img = median(noisy_img)
io.imshow(med_img)
```

```python
'''
(3) Gaussion Filters
'''
sigma_8 = gaussian(noisy_img, sigma=8)
sigma_3 = gaussian(noisy_img, sigma=3)
sigma_0_2 = gaussian(noisy_img, sigma=0.2)
show_images([sigma_8, sigma_3, sigma_0_2],["sigma = 8","sigma = 3","sigma = 0.2"])
```
